### 6. Create a new column as 'published_month' using the column published_at (display the months only)

````python
df['published_month']=df['published_at'].dt.month #created a new column 'published_month' 
````

````python
df
````

### 7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on.....

````python
df['published_month'].replace({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',
                              9:'Sep',10:'Oct',11:'Nov',12:'Dec'},inplace=True)
````

````python
df['published_month'].to_frame()
````

### 8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count.

````python
month=df.groupby('published_month')['video_id'].count().sort_values(ascending=False)
print(month)
````

````python
pd.pivot_table(df,index='published_month',values='video_id',aggfunc='count').sort_values('video_id',ascending=False)
````

### 9. Find the count of unique video_id, channel_id and channel_title.

````python
video_id_unique_count=df['video_id'].nunique() # nunique() counts number of distinct elements 
channel_id_unique_count=df['channel_id'].nunique()
channel_title_unique_count=df['channel_title'].nunique() 
print('The count of unique video_id is:',video_id_unique_count)
print('The count of unique channel_id is:',channel_id_unique_count)
print('The count of unique channel_title is:',channel_title_unique_count)
````

### 10. Find the top10 channel names having the highest number of videos in the dataset and the bottom10having lowest number of videos.

## TOP 10

````python
df.groupby(by="channel_title")['video_id'].count().sort_values(ascending=False).head(10).to_frame() # top 10
````

````python
pd.pivot_table(df,index="channel_title",values='video_id',aggfunc="count").sort_values(by='video_id',ascending=False).head(10) # top 10
````

## BOTTOM 10

````python
df.groupby(by="channel_title")['video_id'].count().sort_values(ascending=False).tail(10).to_frame() # bottom 10
````

````python
pd.pivot_table(df,index="channel_title",values='video_id',aggfunc="count").sort_values(by='video_id',ascending=False).tail(10) # bottom 10
````

### 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

````python
df[['view_count','dislikes']].corr()
````

````python
plt.scatter(df['view_count'],df['dislikes'])
plt.title('scatter plot view_count VS dislikes')
plt.xlabel('view_count')
plt.ylabel('dislikes')
plt.show()
````

```INFERENCE:-```

```
From the graph,we can conclude that the a positive relationship between each other.If one variable increases,other variable also increases.From the correlation table, the correlation between view_count and dislikes is 0.684 so it is positive correlation between each other.
```
